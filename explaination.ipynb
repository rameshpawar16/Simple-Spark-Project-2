Step 1 – Create Spark Session
A Spark session is created to initialize the Spark application and allow distributed data processing.

Step 2 – Load CSV Dataset
The CSV dataset is loaded into a Spark DataFrame using spark.read.csv() with header and schema inference.

Step 3 – Handle Missing Values
Missing values in the dataset are replaced using the fillna() function to ensure the data is clean for machine learning.

Step 4 – Feature Selection
Relevant columns such as age, marks, and attendance are selected as input features for the machine learning model.

Step 5 – Convert to ML-Ready Format
The VectorAssembler from Spark ML is used to combine selected features into a single feature vector required by ML algorithms.

Step 6 – Save Processed Dataset
The processed dataset is saved as a CSV file which can be used for training machine learning models.
